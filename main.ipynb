```python
import pandas as pd
import numpy as np
from src.lstm_step_detect import detect_steps_with_lstm
from src.heading import calculate_path
from src.path_plot import plot_path
from src.load_csv import load_and_preprocess_csv
from src.preprocessing import filter_data

# 1. 데이터 로드 및 전처리
sensor_data_path = 'data/test/sensor_data_2.csv'
df = load_and_preprocess_csv(sensor_data_path)
df = filter_data(df)

# 2. 걸음 검출
features = ['Filtered_Acc_x', 'Filtered_Acc_y', 'Filtered_Acc_z', 
                'Filtered_Gyro_x', 'Filtered_Gyro_y', 'Filtered_Gyro_z', 
                'Filtered_AccNorm']
model_path = 'models/100LSTM_APD_20240914_033745.h5'
step_points = detect_steps_with_lstm(model_path, df, features)

# 3. 경로 계산
gyro_data = df[['Gyroscope x', 'Gyroscope y', 'Gyroscope z']].values
initial_orientation = df[['Oriention x', 'Oriention y', 'Oriention z']].iloc[0].values
predicted_path = calculate_path(gyro_data, step_points, initial_orientation)

# 4. 실제 경로 정의
true_x_positions = [-10.5, 10, 10, -10.5, -10.5]
true_y_positions = [0, 0, 14, 14, 0]
true_path = np.column_stack((true_x_positions, true_y_positions))

# 5. 경로 시각화
plot_path(true_path, predicted_path)




```
